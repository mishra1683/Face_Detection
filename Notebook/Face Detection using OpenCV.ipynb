# Importing required Libraries


```python
import cv2
from random import randrange
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the classifier for Frontal Face 

```python
face_data=cv2.CascadeClassifier(r'C:\Users\hp\Desktop\Project\Face Detection using OpenCV\haarcascade_frontalface_default.xml')
```

# Testing 

```python
image=cv2.imread(r'C:\Users\hp\Desktop\Project\Face Detection using OpenCV\Robert-Downey-Jr.jpg')
cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
coordinate=face_data.detectMultiScale(image)
(x,y,w,h)=coordinate[0]
cv2.rectangle(image,(x,y),((x+w),(y+h)),(0,255,0),3)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

# Testing on Group Photo

```python
image=cv2.imread(r'C:\Users\hp\Desktop\Project\Face Detection using OpenCV\uwMr805gnY.jpg')
cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
coordinate=face_data.detectMultiScale(image)
for (x,y,w,h) in coordinate:
    cv2.rectangle(image,(x,y),((x+w),(y+h)),(randrange(155,256),randrange(155,256),randrange(155,256)),4)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

# Real time Face Detection

```python
video=cv2.VideoCapture(0)
while True:
    rev,frame=video.read()
    frame=cv2.flip(frame,1,1)q
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    coordinate=face_data.detectMultiScale(gray)
    for (x,y,w,h) in coordinate:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),1)
        cv2.putText(frame,"face",(x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)
    cv2.imshow('image',frame)
    key=cv2.waitKey(1)
    if key== ord("q"):
        breakq
video.release()
cv2.destroyAllWindows()
```
